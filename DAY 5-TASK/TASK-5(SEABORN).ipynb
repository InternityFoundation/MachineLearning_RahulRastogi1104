## Importing seaborn packages

```python
import seaborn as sns
sns.set()
```

### loading "tips" dataset into the variable tips

```python
tips = sns.load_dataset('tips')
```

### Dataset "tips" description

```python
tips.head()
```

```python
tips.describe()
```

```python
tips.columns
```

```python
bill = tips['total_bill']
```

```python
sns.distplot(bill , kde=True , bins=50)
```

```python
sns.jointplot(data=tips, x='total_bill' , y='tip', kind='scatter')
```

```python
sns.jointplot(data=tips, x='total_bill' , y='tip', kind='hex')
```

```python
sns.jointplot(data=tips, x='total_bill' , y='tip', kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.barplot(data=tips, x='sex' ,  y='total_bill' )
```

```python
sns.countplot(data=tips , x='sex')
```

```python
sns.boxplot(data=tips , x='day', y='total_bill', palette='rainbow')
```

```python
sns.violinplot(data=tips , x='day', y='total_bill', palette='rainbow')
```

```python
sns.stripplot(data=tips , x='day', y='total_bill', palette='Set1')
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr() , annot=True)
```
