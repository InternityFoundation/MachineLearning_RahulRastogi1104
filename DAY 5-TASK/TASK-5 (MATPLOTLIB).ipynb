```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('SalaryData.csv')
```

```python
dataset.head()
```

```python
X=dataset['YearsExperience']
```

```python
y=dataset['Salary']
```

### Ploting simple graph

```python
plt.plot(X,y)
plt.title("Ist graph")
plt.show()
```

```python
plt.plot(X,y)
plt.title("Modified version of Ist graph")
plt.xlabel('YearsExperience')
plt.ylabel("Salary")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(X,y,"go")
plt.xlabel('YearsExperience')
plt.ylabel("Salary")
plt.title("we use go for green circles")

plt.subplot(1,2,2)
plt.plot(X,y,"r^")
plt.xlabel('YearsExperience')
plt.title("we use r^ for red triangles")


plt.suptitle("My sub plots")
plt.show()
```

### Bar graph

```python
divisions=["Div-A","Div-B","Div-C","Div-D"]
avg_marks=[70,85,68,73]
plt.xlabel('Division')
plt.ylabel("Average marks")
plt.title("BAR GRAPH")
plt.bar(divisions,avg_marks,color='red')
```

### PIE CHART

```python
divisions=["Div-A","Div-B","Div-C","Div-D"]
avg_marks=[70,85,68,73]
Explode=[0,0.1,0,0.1]
plt.title("PIE ChART")
plt.pie(avg_marks,explode=Explode,labels=divisions,shadow=True,startangle=45)
plt.axis()
plt.show()
```

### Histogram

```python
plt.hist(X,20)
plt.xlabel('YearsExperience')
plt.title("HISTOGRAM")
plt.show()
```

### Scatter Graph

```python
plt.scatter(X,y)
plt.xlabel('YearsExperience')
plt.ylabel("Salary")
plt.title("SCATTER GRAPH")
plt.show
```
